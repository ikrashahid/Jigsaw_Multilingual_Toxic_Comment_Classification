# Part 1 - Baseline Classifier

Train DistilBERT on stratified Jigsaw subsets and report required metrics.

```python
import numpy as np
from sklearn.metrics import confusion_matrix
from assignment_workflow import (
    load_and_split_data,
    train_distilbert,
    tokenize_df,
    predict_proba,
    evaluate_binary,
    threshold_sweep,
    plot_roc_pr,
)

train_df, eval_df = load_and_split_data('jigsaw-unintended-bias-train.csv')
trainer, tokenizer = train_distilbert(train_df, eval_df, output_dir='saved_model/part1_baseline')
```

```python
eval_dataset = tokenize_df(eval_df, tokenizer)
y_true = eval_df['label'].values
y_prob = predict_proba(trainer, eval_dataset)

metrics = evaluate_binary(y_true, y_prob, threshold=0.5)
print('Baseline metrics:', metrics)
print('Confusion matrix @0.5:\n', confusion_matrix(y_true, (y_prob >= 0.5).astype(int)))
print('\nThreshold sweep:')
print(threshold_sweep(y_true, y_prob, [0.3, 0.4, 0.5, 0.6, 0.7]))
plot_roc_pr(y_true, y_prob)
```
